#### Demo for getting historical return data

#### Historical constituents info from https://github.com/fja05680/sp500 ; Risk free data from https://mba.tuck.dartmouth.edu/pages/faculty/ken.french/data_library.html

```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
name_path = r"./sp500-master/sp500-master/S&P 500 Historical Components & Changes(12-30-2023).csv"
name_data = pd.read_csv(name_path)
name_data["date"] = pd.to_datetime(name_data["date"], format="%Y-%m-%d")
name_data = name_data.set_index("date")
name_data = name_data["2003-10-01" : "2023-10-30"]

rf_path = "./F-F_Research_Data_Factors_daily_CSV/F-F_Research_Data_Factors_daily.csv"
rf_data = pd.read_csv(rf_path, header=3)
rf_data = rf_data.iloc[:-1,:]
rf_data.set_index(pd.to_datetime(rf_data.iloc[:,0], format="%Y%m%d"), inplace=True)
rf_data = rf_data.iloc[:,1:]
rf_data.index.name = "date"

display(name_data, rf_data)
```

```python
name_srs = pd.Series(name_data.sum().iloc[0].split(","))
name_to_filter = name_srs.value_counts()
name_arr = np.array(name_to_filter.loc[name_to_filter == name_data.shape[0]].index.str.replace(".","-"))
name_arr
```

```python
name_df = pd.DataFrame()
for i in range(name_arr.shape[0]):
    temp = yf.Ticker(name_arr[i])
    name_df = pd.concat([name_df,temp.history(start="2003-10-01", end="2023-10-30")["Close"]], axis=1)
name_df.columns = name_arr
name_df.reset_index(inplace=True)
name_df["index"] = pd.to_datetime(name_df["index"].dt.strftime("%Y-%m-%d"), format="%Y-%m-%d")
name_df.set_index("index",drop=True,inplace=True)
name_df
```

```python
return_df = (name_df / name_df.shift(1) - 1).dropna()
return_df = return_df.join((rf_data["RF"] / 100).rename("risk_free"))
stock_return = return_df.iloc[:,:-1]
risk_free = return_df.iloc[:,-1]
excess_return = stock_return.apply(lambda x : x - risk_free)
excess_return
```

```python
excess_return.to_csv("./return_data.csv")
```
